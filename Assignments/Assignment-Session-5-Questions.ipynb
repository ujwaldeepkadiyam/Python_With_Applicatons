### **Step 1: Melt the DataFrame**

We convert the score columns into a long format:

````python
score_cols = ['SAS','R','PYTHON','SQL','EXCEL']

df_melted = df.melt(id_vars=['NAME'], value_vars=score_cols,
                    var_name='SUBJECT', value_name='SCORE')

df_melted.head()
````

#### Q9: compute **average scores per group** using `melt` + `groupby`. Here‚Äôs how:

### **Step 1: Melt the DataFrame (scores only)**

#### Q10:  Use **`groupby().transform()`**, to calculate **group-level averages** **directly** without melting or merging. 

- `transform` returns a Series aligned with the original DataFrame, so it can be assigned as a new column.

### **1Ô∏è‚É£ Row-wise average (per student)**

````python
score_cols = ['SAS','R','PYTHON','SQL','EXCEL']

# Average score per row
df['AVERAGE_SCORE'] = df[score_cols].mean(axis=1)
df[['NAME','AVERAGE_SCORE']]
````

‚úÖ Same as before, simple row-wise mean.


### **2Ô∏è‚É£ Average score per GENDER using transform**

````python
# Compute average of all score columns per gender
df['GENDER_AVG'] = df[score_cols].mean(axis=1)  # start with row-wise sum
df['GENDER_AVG'] = df.groupby('GENDER')[score_cols].transform('mean').mean(axis=1)

df[['NAME','GENDER','GENDER_AVG']]
````

**Explanation:**

1. `df.groupby('GENDER')[score_cols]` ‚Üí select numeric columns per group.
2. `.transform('mean')` ‚Üí compute **mean per column within the group**, returned as a DataFrame aligned with `df`.
3. `.mean(axis=1)` ‚Üí average across the score columns per student to get **group average**.


### **3Ô∏è‚É£ Average score per EDUCATION using transform**

````python
df['EDU_AVG'] = df.groupby('EDUCATION')[score_cols].transform('mean').mean(axis=1)
df[['NAME','EDUCATION','EDU_AVG']]
````

### ‚úÖ **Advantages of using transform**

* No need to `melt` or `merge`.
* Returns a **Series aligned with original DataFrame**.
* Works nicely for **row-level + group-level computations**.

# üìù Assignment: Radar Chart with Custom Parameters

### Dataset & Code

````python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

# New dataset
df = pd.DataFrame({
    'group': ['X','Y','Z','W'],
    'speed': [25, 35, 15, 30],
    'accuracy': [40, 20, 35, 25],
    'strength': [30, 10, 25, 40],
    'stamina': [15, 35, 20, 30],
    'strategy': [20, 40, 25, 15]
})

#  ---- PART 1: Create background
categories = list(df)[1:]   # variable names
N = len(categories)

# Angle calculation
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# Initialise radar chart
ax = plt.subplot(111, polar=True)

# Question: Why do we set offset and direction here?
ax.set_theta_offset(pi / 2)       # places first axis at the top
ax.set_theta_direction(-1)        # clockwise

# Add labels
plt.xticks(angles[:-1], categories)

# Question: What happens if we change ticks and limits?
ax.set_rlabel_position(0)
plt.yticks([10,20,30,40], ["10","20","30","40"], color="grey", size=7)
plt.ylim(0,50)   # setting limit higher than max to give padding

# ------- PART 2: Plot data
values = df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=2, linestyle='solid', label="Group X")
ax.fill(angles, values, 'b', alpha=0.2)

# Add more groups for comparison
for i in range(1, len(df)):
    values = df.loc[i].drop('group').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=f"Group {df.loc[i,'group']}")
    ax.fill(angles, values, alpha=0.1)

# Add title and legend
plt.title("Radar Chart - Player Performance", size=14, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(0.1,0.1))

plt.show()
````

````python
# üìä Visualization with Stacked Bar Chart

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define categories and values for each bar segment
departments = ['HR', 'Finance', 'IT', 'Marketing']

salaries = [40, 55, 60, 35]   # Base salary component
bonuses = [10, 15, 12, 8]     # Bonus component
benefits = [5, 10, 7, 6]      # Benefits component

# Step 2: Set width of each bar and compute x-axis positions
bar_width = 0.5
index = np.arange(len(departments))

# Step 3: Plot salaries (bottom stack)
plt.bar(index, salaries, bar_width, label='Salaries')

# Step 4: Plot bonuses on top of salaries
plt.bar(index, bonuses, bar_width, bottom=salaries, label='Bonuses')

# Step 5: Plot benefits on top of (salaries + bonuses)
plt.bar(index, benefits, bar_width, 
        bottom=np.array(salaries) + np.array(bonuses), 
        label='Benefits')

# Step 6: Customize chart
plt.xlabel('Departments')
plt.ylabel('Total Compensation')
plt.title('Department-wise Compensation Breakdown')
plt.xticks(index, departments)
plt.legend()

# Step 7: Show chart
plt.show()
````

# üìù Assignment: Scatter Plot Visualization

### **Q1.**

Create a scatter plot using **Matplotlib** and **Seaborn** for the following dataset. Use different colors for categories in Seaborn.

````python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create a NEW dataset
data = {
    'X': [10, 15, 8, 20, 5, 18, 12, 25, 7, 14],
    'Y': [50, 60, 45, 70, 40, 65, 55, 80, 42, 58],
    'Category': ['Red', 'Blue', 'Red', 'Blue', 'Red', 'Blue', 'Red', 'Blue', 'Red', 'Blue']
}

df = pd.DataFrame(data)
````

````python
# Step 3: Scatterplot using Matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], c='green', label='Data points', s=80)
plt.title('Scatterplot using Matplotlib')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
````

````python
# Step 4: Scatterplot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', hue='Category', data=df, palette='coolwarm', s=100)
plt.title('Scatterplot using Seaborn')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
````

### ‚úÖ Solution / Explanation

1. **Matplotlib Scatterplot**

   * `plt.scatter(df['X'], df['Y'], c='green', s=80)` plots all points in **green** with size 80.
   * `plt.legend()` adds a label.
   * `plt.title()`, `plt.xlabel()`, `plt.ylabel()` label the chart.

2. **Seaborn Scatterplot**

   * `sns.scatterplot(x='X', y='Y', hue='Category', data=df, palette='coolwarm', s=100)` automatically colors points by `Category`.
   * `s=100` controls marker size.
   * `palette='coolwarm'` provides a visually distinct color mapping for categories.

3. **Observation / Insight**

   * Points can be visually distinguished based on the `Category` variable.
   * Matplotlib requires manual coloring, while Seaborn provides **automatic hue mapping** for categorical variables.




# üìù Assignment: Bubble Chart Visualization

### **Q1.**

Create a bubble chart using Matplotlib for the given dataset. Use `X` for x-axis, `Y` for y-axis, and `Size` for bubble sizes. Include a legend showing the bubble size scale.

````python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create a NEW dataset
data = {
    'X': [10, 15, 12, 20, 8, 25, 18, 30, 22, 16],  # X-axis values
    'Y': [60, 75, 70, 85, 55, 90, 80, 95, 78, 72], # Y-axis values
    'Size': [150, 250, 200, 300, 180, 350, 280, 400, 220, 260],  # Bubble sizes
    'Category': ['Red', 'Blue', 'Red', 'Blue', 'Red', 'Blue', 'Red', 'Blue', 'Red', 'Blue']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Step 3: Plot the bubble chart
plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    df['X'], df['Y'], 
    s=df['Size'],          # Bubble sizes
    alpha=0.5,             # Semi-transparent
    c='orange',            # Bubble color
    edgecolors='w',        # White edge
    label='Data points'
)

# Add title and labels
plt.title('Bubble Chart - Department Metrics')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Add legend for bubble sizes
handles, labels = scatter.legend_elements(prop='sizes', alpha=0.6)
plt.legend(handles, labels, loc="upper right", title="Bubble Sizes")

# Show plot
plt.show()
````

### ‚úÖ Solution / Explanation

1. **DataFrame creation**

   * `df = pd.DataFrame(data)` creates a tabular structure for plotting.

2. **Bubble Chart using Matplotlib**

   * `s=df['Size']` ‚Üí controls bubble area according to the `Size` column.
   * `alpha=0.5` ‚Üí makes bubbles semi-transparent.
   * `c='orange'` ‚Üí sets bubble color.
   * `edgecolors='w'` ‚Üí adds a white border for better visibility.

3. **Legend for bubble sizes**

   * `scatter.legend_elements(prop='sizes', alpha=0.6)` automatically generates bubble-size legend.
   * `plt.legend()` displays the legend on the chart.

4. **Observation / Insight**

   * Larger bubbles represent higher values in the `Size` column.
   * Plot provides a visual representation of **X vs Y metrics with magnitude** indicated by bubble size.

# üìù Assignment: Histogram Visualization

### **Q1.**

Create a histogram to visualize the distribution of the dataset using **Matplotlib** and **Seaborn**. Use appropriate bins and colors, and add a **KDE curve** in Seaborn.

````python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create a NEW dataset
data = {
    'Values': [15, 20, 25, 18, 30, 22, 40, 35, 28, 32, 45, 38, 50, 42, 60, 55, 48, 70, 65, 58]
}

df = pd.DataFrame(data)
````

````python
# Step 3: Plot histogram using Matplotlib
plt.figure(figsize=(8, 6))
plt.hist(df['Values'], bins=8, color='lightgreen', edgecolor='black')
plt.title('Histogram using Matplotlib')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
````

````python
# Step 4: Plot histogram using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(df['Values'], bins=8, kde=True, color='lightgreen')
plt.title('Histogram using Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
````

### ‚úÖ Solution / Explanation

1. **Matplotlib Histogram**

   * `plt.hist(df['Values'], bins=8, color='lightgreen', edgecolor='black')` divides the data into 8 equal-width intervals.
   * `color='lightgreen'` fills the bars.
   * `edgecolor='black'` outlines each bar.
   * `plt.grid(axis='y', alpha=0.75)` adds horizontal grid lines.

2. **Seaborn Histogram**

   * `sns.histplot(df['Values'], bins=8, kde=True, color='lightgreen')` adds a **KDE curve** showing the density estimate.
   * Automatically colors bars according to the `color` parameter.

3. **Purpose / Insight**

   * Both charts **show the frequency distribution** of numerical values.
   * Helps identify **concentration, spread, and skewness** in the dataset.



# üìù Assignment: Box Plot Visualization

### **Q1.**

Create a box plot using **Matplotlib** to visualize the distribution of values for each category in the dataset. Customize the box color and add grid lines for clarity.

````python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create a NEW dataset
data = {
    'Category': ['X', 'X', 'X', 'X', 'X', 
                 'Y', 'Y', 'Y', 'Y', 'Y', 
                 'Z', 'Z', 'Z', 'Z', 'Z'],
    'Values': [12, 15, 14, 13, 16, 
               22, 25, 23, 21, 24, 
               35, 38, 36, 37, 34]
}

df = pd.DataFrame(data)
````

````python
# Step 3: Box plot using Matplotlib
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['Category'] == 'X']['Values'],
             df[df['Category'] == 'Y']['Values'],
             df[df['Category'] == 'Z']['Values']],
            labels=['X', 'Y', 'Z'],
            patch_artist=True,
            boxprops=dict(facecolor='lightcoral'))
plt.title('Box Plot using Matplotlib')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
````

### ‚úÖ Solution / Explanation

1. **DataFrame creation**

   * `df = pd.DataFrame(data)` stores the data in a tabular format.
   * `Category` distinguishes groups, `Values` contains numerical data.

2. **Box Plot using Matplotlib**

   * `plt.boxplot([...], labels=['X','Y','Z'], patch_artist=True, boxprops=dict(facecolor='lightcoral'))`

     * `patch_artist=True` allows filling boxes with color.
     * `boxprops=dict(facecolor='lightcoral')` sets the box color.
   * Each list inside `plt.boxplot()` corresponds to one category.

3. **Grid and labels**

   * `plt.grid(axis='y')` adds horizontal grid lines.
   * `plt.title()`, `plt.xlabel()`, `plt.ylabel()` label the plot.

4. **Purpose / Insight**

   * Box plots visualize **median, quartiles, range, and potential outliers** for each category.
   * Helps compare the distribution between different groups.



# üìù Assignment: Line Chart Visualization

### **Q1.**

Create a line chart to compare two variables over time using both **Matplotlib** and **Seaborn**. Include markers at each point, rotate x-axis labels for better readability, and add a legend.

````python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create a NEW dataset
data = {
    'Date': pd.date_range(start='2025-08-01', periods=10, freq='D'),
    'Sales_A': [120, 135, 150, 145, 160, 155, 170, 165, 180, 175],
    'Sales_B': [100, 110, 115, 120, 125, 130, 140, 135, 145, 150]
}

df = pd.DataFrame(data)

# Step 3: Line chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sales_A'], marker='o', linestyle='-', color='b', label='Sales A')
plt.plot(df['Date'], df['Sales_B'], marker='o', linestyle='-', color='r', label='Sales B')
plt.title('Line Chart using Matplotlib')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Line chart using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Sales_A', data=df, marker='o', color='b', label='Sales A')
sns.lineplot(x='Date', y='Sales_B', data=df, marker='o', color='r', label='Sales B')
plt.title('Line Chart using Seaborn')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
````

### ‚úÖ Solution / Explanation

1. **DataFrame creation**

   * `df = pd.DataFrame(data)` stores the time series data.
   * `Date` column represents time; `Sales_A` and `Sales_B` are numerical values.

2. **Matplotlib line chart**

   * `plt.plot()` draws lines connecting data points with markers (`marker='o'`).
   * `linestyle='-'` connects points with lines.
   * `color` differentiates the two lines.
   * `plt.legend()` labels the lines.
   * `plt.xticks(rotation=45)` improves readability of date labels.

3. **Seaborn line chart**

   * `sns.lineplot()` plots each series with markers.
   * Automatically handles legends and aesthetics.

4. **Purpose / Insight**

   * Line charts visualize **trends over time**.
   * Comparison of two series shows which variable grows faster or has higher values over the same period.



# üìù Assignment: Heatmap Visualization

### **Q1.**

Create a heatmap using Seaborn for weekly performance data. Annotate each cell, customize the color map, and include grid lines.

````python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Create a NEW dataset
data = {
    'Monday': [12, 18, 25, 30],
    'Tuesday': [15, 22, 28, 35],
    'Wednesday': [20, 26, 33, 40],
    'Thursday': [25, 30, 38, 45],
    'Friday': [28, 35, 42, 50]
}

index = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
df = pd.DataFrame(data, index=index)

# Step 3: Plot heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', linewidths=.5, linecolor='gray')
plt.title('Weekly Performance Heatmap')
plt.xlabel('Day')
plt.ylabel('Week')
plt.show()
````

### ‚úÖ Solution / Explanation

1. **DataFrame creation**

   * `df = pd.DataFrame(data, index=index)` stores weekly performance values.
   * Rows represent weeks; columns represent weekdays.

2. **Seaborn heatmap**

   * `sns.heatmap(df, annot=True, cmap='coolwarm', linewidths=.5, linecolor='gray')`

     * `annot=True` adds the numerical values in each cell.
     * `cmap='coolwarm'` provides a visually distinct gradient.
     * `linewidths=.5` and `linecolor='gray'` draw grid lines between cells.

3. **Purpose / Insight**

   * Heatmaps visualize **patterns, trends, and differences** across categories.
   * Allows quick identification of higher and lower values for weekdays across weeks.
